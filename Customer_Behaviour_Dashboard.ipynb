```python
import pandas as pd
import plotly.express as px

# Ensure df is defined and preprocessed
df = pd.read_csv("/content/customer_shopping_behavior.csv")
df.columns = df.columns.str.lower().str.replace(" ", "_").str.replace(r"[()]", "", regex=True)
df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 18, 30, 45, 60, 100],
    labels=["<18", "18-30", "31-45", "46-60", "60+"]
)

fig = px.bar(
    df["age_group"].value_counts().reset_index(),
    x="age_group",
    y="count",
    title="Customer Age Groups",
    labels={"age_group": "Age Group", "count": "Number of Customers"}
)
fig.show()
```

```python
category_sales = df.groupby("category")["purchase_amount_usd"].sum().reset_index()

fig = px.bar(
    category_sales,
    x="category",
    y="purchase_amount_usd",
    title="Total Purchase Amount by Category",
    color="category"
)
fig.show()
```

```python
fig = px.pie(
    df,
    names="payment_method",
    title="Payment Method Distribution"
)
fig.show()
```

```python
fig = px.histogram(
    df,
    x="subscription_status",
    title="Subscription Status Distribution",
    color="subscription_status"
)
fig.show()
```

```python
fig = px.box(
    df,
    x="discount_applied",
    y="purchase_amount_usd",
    title="Discount Applied vs Purchase Amount"
)
fig.show()
```

```python
top_locations = df["location"].value_counts().head(10).reset_index()
top_locations.columns = ["location", "count"]

fig = px.bar(
    top_locations,
    x="location",
    y="count",
    title="Top 10 Customer Locations"
)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="frequency_of_purchases",
    y="purchase_amount_usd",
    color="gender",
    title="Purchase Frequency vs Purchase Amount"
)
fig.show()
```
